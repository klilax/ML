```python
import matplotlib.pyplot as plt
# import matplotlib_inline.backend_inline as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
x = np.linspace(0, 10, 200)  # x data
y = np.sin(x) + x            # y data
ax.plot(x, y)                # creates the plot
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
fig.suptitle('Our beautiful plot', fontsize=40)
color = 'y'
linestyle = 'dotted'
linewidth = 4

ax.set_ylabel('function', fontsize = 30)
ax.set_xlabel('argument', fontsize = 30)
ax.
for xtick in ax.xaxis.get_major_ticks():
    xtick.label.set_fontsize(18)

for ytick in ax.yaxis.get_major_ticks():
    ytick.label.set_fontsize(18)

ax.plot(x, y, c=color, linestyle=linestyle, linewidth=linewidth)
```

```python
plt.close(fig)
```
